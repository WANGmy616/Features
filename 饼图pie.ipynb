```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
oses = {
    'windows7':60.86,
    'windows10':18.46,
    'windows8':3.61,
    'windows xp':10.3,
    'mac os':6.78,
    '其他':1.12
    
     }
plt.title('pie')
patches,texts,autotexts= plt.pie(oses.values(),autopct='%.2f%%',labels=oses.keys(),shadow=True,explode=(0,0.1,0,0,0,0))
for autotext in autotexts:
    autotext.set_color('w')
    autotext.set_size(8)
plt.show()
```
