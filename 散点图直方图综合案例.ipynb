```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
import matplotlib.ticker as ticker #设置刻度
```

```python
fig,axes = plt.subplots(1,2)
axes[1].set_position([0.2,0.2,0.4,0.4])
```

```python
midwest = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/midwest_filter.csv")
popasian = midwest[midwest['inmetro']==1][['popamerindian','popasian']]
fig = plt.figure(figsize=(10,10))

widths = (4,1)
heights = (1,4)

gs = fig.add_gridspec(2,2,width_ratios=widths,height_ratios=heights)
ax1 = fig.add_subplot(gs[0,0])
ax1.hist(popasian['popamerindian'])
for tick in ax1.xaxis.get_major_ticks():
    tick.label1.set_visible(False)
    
    
ax2 = fig.add_subplot(gs[1,0])
ax2.scatter('popamerindian','popasian',data=popasian)


ax3 = fig.add_subplot(gs[1,1])
ax3.hist(popasian['popasian'],orientation='horizontal')
for ticks in ax3.yaxis.get_major_ticks():
    ticks.label1.set_visible(False)
    
midwest.head(100)
fig.tight_layout( )
```
