```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# tips = sns.load_dataset('tips',cache=True)
tips = pd.read_csv('dataset/tips.csv')
tips.head()
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',col='time',row='sex',data=tips,size='size')#row = 'time' 按照行排列
```

```python
sns.relplot(x='total_bill',y='tip',col='day',col_wrap=2,data=tips)
```

# markdown
1.sns.relplot(kind='scatter') , sns.scatterplot() 一样
2.sns.relplot(kind='line'),sns.lineplot()一样
3.relplot在绘制的时候，不能直接给x,y指定具体的值，而应该使用DataFrame中的列的名字
4.hue表示的是颜色，应该制定某列的名字，那么sns会自动的将指定的列的 值的个数，取不同的颜色
5.col表示分成几个图像，指定列的名字
5.在绘制折线时，可以指定线条样式 style
