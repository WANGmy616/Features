```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
duration = [1,4,3,5,6,7,8,9,0,3,4,5,6,7,8,9,2,3,5,7,6,8,0,2,2,3,8]
plt.figure(figsize=(15,10))
nums,bins,patches = plt.hist(duration,bins=10
                             ,edgecolor='k')#cumulative=True累积相加 bin是分成几块来计数的

plt.xticks(bins,bins)#第一个真正的刻度，第一个显示名字
for num,bin in zip(nums,bins):
    plt.annotate('%.4f'%num,xy=(bin,num),xytext=(bin+0.2,num+0.001))
plt.legend()  
plt.show()
```
