```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
planets = sns.load_dataset('planets')
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x='year',y='distance',data=planets,ci=None,palette=sns.color_palette('dark',n_colors=23))
```

```python
current_color = sns.color_palette('hls',n_colors=20)
sns.palplot(current_color)
sns.palplot(sns.color_palette('Paired'))


sns.palplot(sns.hls_palette(10,h=0.4,l=0.4,s=1))#饱和度

sns.palplot(sns.husl_palette(10))


```

```python
sns.set_palette('hls')
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
sinplot()
```
