```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
import matplotlib.ticker as ticker #设置刻度
```

```python
#调整拥挤 图之间的间距等
def example_plot(ax,fontsize=12):
    ax.plot([1,2])
    ax.set_xlabel('x-label',fontsize=fontsize)
    ax.set_ylabel('y-label',fontsize=fontsize)
    ax.set_title('Title',fontsize=fontsize)
    
fig,axes = plt.subplots(2,2)
fig.set_facecolor('y')

example_plot(axes[0,0])
example_plot(axes[0,1])
example_plot(axes[1,0])
example_plot(axes[1,1])

fig.tight_layout(w_pad=10,h_pad=10)

# fig.subplots_adjust(0,0,1,1,wspace=0.5,hspace=1)
```

```python
fig = plt.figure()
gs = fig.add_gridspec(3,3)
fig.add_subplot(gs[0,0:3])
fig.add_subplot(gs[1,0:2])
fig.add_subplot(gs[1:3,2])
fig.add_subplot(gs[2,0])
fig.add_subplot(gs[2,1])
fig.tight_layout()
```

```python
fig = plt.figure()
widths = (1,2,1)
heights = (3,3,1)
gs = fig.add_gridspec(3,3,width_ratios=widths,height_ratios=heights)
for row in range(0,3):
    for col in range(0,3):
        fig.add_subplot(gs[row,col])
fig.tight_layout()
```
