```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
```

```python
tips = sns.load_dataset('tips')
axes=sns.scatterplot(x='total_bill',y='tip',data=tips)
axes.set_xticks(range(0,60,2))
```

```python
fig,[ax1,ax2]=plt.subplots(1,2,figsize=(20,5))
sns.scatterplot(x='total_bill',y='tip',data=tips,ax=ax1)
sns.barplot(x='day',y='total_bill',data=tips,ax=ax2)
```

```python
g = sns.FacetGrid(tips)
g.map(plt.scatter,'total_bill','tip')
```

```python

sns.set(rc={'font.sans-serif':'HeiTi TC'})
g = sns.FacetGrid(tips,col="day",col_wrap=2,hue='time')
g.map(sns.regplot,"total_bill","tip")
# g.add_legend(title='中文',labels=['一','er'],label_order=['lunch','dinner'])
new_labels = ['午餐','晚餐']
g.add_legend(title="时间")
for t,l in zip(g._legend.texts,new_labels):
    t.set_text(l)
```

```python
#设置标题
g = sns.FacetGrid(tips,col="day",row="time")
g.map(sns.regplot,"total_bill","tip")
g.set_titles(template="时间{row_name}/星期{col_name}")
#设置坐标轴
g.set_axis_labels('消费金额','小费')
g.set(facecolor='r')
g.fig.set_dpi(100)
```
