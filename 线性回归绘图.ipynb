```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.regplot(tips['total_bill'],tips['tip'])
```

 # 笔记:
1. regplot和lmplot都可以绘制线性回归图。
2. regplot的x和y可以为numpy的数组或者是Series对象。
3. 一些其他的参数，比如order，robust等
